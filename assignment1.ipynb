-----------------------------------------------------------------------------------------------

Find each company's Highest Price Car

```python
df.groupby('Company').agg({'Price': ['max']})
```

-----------------------------------------------------------------------------------------------

Find the Avg Mileage of each car making company.

```python
df1 = df.groupby('Company').agg({'City_Mpg': ['mean']})
df2 = df.groupby('Company').agg({'Highway_Mpg': ['mean']})
pd.merge(df1, df2, on = "Company", how = "inner")
```
